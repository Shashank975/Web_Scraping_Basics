# **Final Code for Scraping**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm  # Import tqdm for progress bar

# Initialize an empty list to store DataFrames
dfs = []

# Define the total number of pages to scrape
total_pages = 500

# Initialize tqdm with total number of pages
with tqdm(total=total_pages, desc='Scraping Pages', unit='page') as pbar:
    # Iterate over pages with tqdm progress bar
    for k in range(1, total_pages + 1):
        url = f'https://www.ambitionbox.com/list-of-companies?page={k}'

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
        }

        try:
            # Fetch webpage content
            webpage = requests.get(url, headers=headers).text

            # Parse HTML using BeautifulSoup
            soup = BeautifulSoup(webpage, 'html.parser')

            # Find company and reviews elements
            company = soup.find_all('div', class_='companyCardWrapper__metaInformation')
            reviews = soup.find_all('div', class_='companyCardWrapper__tertiaryInformation')

            # Initialize lists to store the extracted data for this page
            name = []
            rating = []
            work_place = []
            adv = []
            disadv = []
            Rview = []
            salary = []
            interview = []
            jobs = []

            # Extract data from company elements
            for i in company:
                # Extract company name
                h2_tag = i.find('h2')
                if h2_tag:
                    name.append(h2_tag.text.strip())
                else:
                    name.append(None)

                # Extract company rating
                rating_tag = i.find('span', class_='companyCardWrapper__companyRatingValue')
                if rating_tag:
                    rating.append(rating_tag.text.strip())
                else:
                    rating.append(None)

                # Extract company work place
                work_place_tag = i.find('div', class_='companyCardWrapper__interLinkingWrapper')
                if work_place_tag:
                    work_place.append(work_place_tag.text.strip())
                else:
                    work_place.append(None)

                # Extract company advantage
                adv_tag = i.find('span', class_='companyCardWrapper__ratingValues')
                if adv_tag:
                    adv.append(adv_tag.text.strip())
                else:
                    adv.append(None)

                # Extract company disadvantage (second span with class 'companyCardWrapper__ratingValues')
                rating_values = i.find_all('span', class_='companyCardWrapper__ratingValues')
                if len(rating_values) > 1:  # Check if there are at least two such elements
                    disadv.append(rating_values[1].text.strip())
                else:
                    disadv.append(None)

            # Extract data from reviews elements
            for j in reviews:
                action_count_tags = j.find_all('span', class_='companyCardWrapper__ActionCount')

                if len(action_count_tags) >= 4:
                    Rview.append(action_count_tags[0].text.strip())  # Reviews
                    salary.append(action_count_tags[1].text.strip())  # Salary
                    interview.append(action_count_tags[2].text.strip())  # Interview
                    jobs.append(action_count_tags[3].text.strip())  # Jobs
                else:
                    Rview.append(None)
                    salary.append(None)
                    interview.append(None)
                    jobs.append(None)

            # Create a dictionary with the extracted data for this page
            data = {
                'Name': name,
                'Rating': rating,
                'Work Place': work_place,
                'Advantage': adv,
                'Disadvantage': disadv,
                'Reviews': Rview,
                'Salary': salary,
                'Interview': interview,
                'Jobs': jobs
            }

            # Convert dictionary to DataFrame
            df = pd.DataFrame(data)

            # Split 'Work Place' column into separate columns
            if 'Work Place' in df.columns:
                split_data = df['Work Place'].str.split('|', expand=True)
                if len(split_data.columns) == 5:  # Ensure the split produced exactly five columns
                    df[['Industry', 'Employees', 'Ownership', 'Age', 'Location']] = split_data
                    df.drop(columns=['Work Place'], inplace=True)  # Drop original 'Work Place' column
                else:
                    print(f"Warning: Unexpected number of columns in 'Work Place' on page {k}. Skipping split operation.")
                    df['Industry'] = None
                    df['Employees'] = None
                    df['Ownership'] = None
                    df['Age'] = None
                    df['Location'] = None

            # Append the DataFrame for this page to the list
            dfs.append(df)

        except Exception as e:
            print(f"Error occurred on page {k}: {e}")
            # Create an empty DataFrame with columns matching expected output
            df = pd.DataFrame(columns=['Name', 'Rating', 'Work Place', 'Advantage', 'Disadvantage',
                                       'Reviews', 'Salary', 'Interview', 'Jobs', 'Industry', 'Employees',
                                       'Ownership', 'Age', 'Location'])
            dfs.append(df)  # Append the empty DataFrame to maintain consistency
            continue  # Continue to the next page even if an error occurs

        finally:
            # Update tqdm progress bar
            pbar.update(1)

# Concatenate all DataFrames in the list into one DataFrame
ambition_box = pd.concat(dfs, ignore_index=True)

# Save the DataFrame to a CSV file
ambition_box.to_csv('ambitionbox_companies.csv', index=False)

# Display the DataFrame
print(ambition_box)
```
