```python
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup 
```

```python
df=pd.read_csv("Quote.csv")
df.head(10)
```

```python
del df["quote"]
del df["author_id"]
del df["tags"]
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df.values
```

```python
"""for author in df.values:
    print (author[0],author[1])"""
author_info=[]
Path=[]
for author in  tqdm (df.values):
    name="Author/"+author[0] + '.txt'
    links=author[1]
    res=requests.get(links)
    soup=BeautifulSoup(res.text,'html.parser')
    text=soup.find('div',class_="author-details").text
    author_info.append(text.strip())
    Path.append(name.strip())
    fd=open(name,'w',encoding="UTF-8")
    fd.write(text)
    fd.close()
    
    
```

```python
df['author_info']=author_info
df["Path"]=Path
```

```python
df.head()
```

```python
df.to_csv("Authorss.csv",index=False)
```
